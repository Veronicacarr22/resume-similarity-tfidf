```python
import pdfplumber
import os
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
import logging
pdf_loggers = [
    logging.getLogger(name)
    for name in logging.root.manager.loggerDict if
name.startswith('pdfminer')
]
for logger in pdf_loggers:
  logger.setLevel(logging.ERROR)
```

```python
import nltk
nltk.download('punkt_tab')
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
stop_words = set(stopwords.words('english'))
```

```python
pdfDF =pd.DataFrame(columns=['FileName','Text'])
pdfFolder = "/content/drive/MyDrive/MSDA683/Module3Data/data/data1"
directory = os.fsencode(pdfFolder)

for file in os.listdir(directory):
  filename = os.fsdecode(file)
  if filename.endswith(".pdf"):
    pdfPath = os.path.join(pdfFolder, filename)
    print(pdfPath)
    with pdfplumber.open(pdfPath) as pdf:
      entirePDF = ""
      for page in pdf.pages:
        text = page.extract_text()
        entirePDF = entirePDF + text
        entirePDF = entirePDF.lower()

      pdfDF.loc[len(pdfDF)] = [filename,entirePDF]
  else:
    continue
```

```python
def remove_stopwords(text):
    if isinstance(text, str):
        word_tokens = word_tokenize(text)
        filtered_text = [word for word in word_tokens if word.lower() not in stop_words]
        return " ".join(filtered_text)
    return ""
```

```python
pdfDF['Text'] = pdfDF['Text'].apply(remove_stopwords)
```

```python
corpus = pdfDF['Text']
vectorizer = TfidfVectorizer(ngram_range=(1,1))
X = vectorizer.fit_transform(corpus)

print(vectorizer.get_feature_names_out())
print(X.toarray())

df = pd.DataFrame(X.toarray(),columns = vectorizer.get_feature_names_out())
print(df)
```

```python
documents = pdfDF['Text']
tfidf = TfidfVectorizer().fit_transform(documents)
pairwise_similarity = tfidf * tfidf.T
pairwise_similarity.toarray()
```

```python
MatrixDF.columns = pdfDF['FileName']
MatrixDF.index = pdfDF['FileName']

MatrixDF
```

```python
idx = 0
FinalDF = pd.DataFrame(columns=['Source File','Match File','Sim Score'])
for doc in MatrixDF:
  DocSimsDF = MatrixDF.iloc[idx].sort_values(ascending=False).head(5)
  # print (doc)
  for filename,score in DocSimsDF.items():
    FinalDF.loc[len(FinalDF)] = [doc,filename,score]
  idx += 1
FinalDF.to_csv('/content/drive/MyDrive/MSDA683/Module3Data/Syllabus/DocSims22.csv', index=False)
FinalDF
```
